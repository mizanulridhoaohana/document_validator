```python
import os
from pdf2image import convert_from_path
```

```python
# Path ke folder dataset
pdf_folder = './dataset'

# Path ke folder output
output_folder = './extracted_data'

# Membuat folder output jika belum ada
os.makedirs(output_folder, exist_ok=True)

# Mendapatkan daftar semua file PDF di folder dataset
pdf_files = [f for f in os.listdir(pdf_folder) if f.endswith('.pdf')]

# Looping melalui setiap file PDF
for pdf_file in pdf_files:
    # Path lengkap ke file PDF
    pdf_path = os.path.join(pdf_folder, pdf_file)

    # Mengonversi halaman-halaman dari PDF menjadi gambar
    pages = convert_from_path(pdf_path)

    # Nama dasar file tanpa ekstensi
    base_filename = os.path.splitext(pdf_file)[0]

    # Looping melalui setiap halaman kecuali halaman pertama
    for i, page in enumerate(pages[1:], start=2):  # Mulai dari halaman kedua
        # Menyusun nama file untuk gambar output
        image_filename = f"{base_filename}_page_{i}.png"
        image_path = os.path.join(output_folder, image_filename)

        # Menyimpan gambar
        page.save(image_path, 'PNG')

    print(f"File {pdf_file} berhasil diekstrak ke {output_folder} (halaman pertama dikecualikan)")
```
